# BAR PLOTS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Bar plot for counts of each label, sorted in descending order
label_counts = df[' Label'].value_counts()
bars=sns.barplot(y=label_counts.index, x=label_counts.values, palette="viridis")

for bar, count in zip(bars.patches, label_counts.values):
    plt.text(count + 0.02 * max(label_counts.values),  # Position annotation slightly to the right of the bar
             bar.get_y() + bar.get_height() / 2,      # Center annotation vertically on the bar
             str(count), 
             ha='left', va='center', fontsize=10)
plt.ylabel('Label', fontsize=12)
plt.xlabel('Count', fontsize=12)
plt.title('Count of Labels', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

# BOX PLOTS

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Remove outliers based on the IQR
def remove_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Filter out outliers from 'Total Fwd Packets'
filtered_df = remove_outliers(df, ' Total Fwd Packets')

# Create the box plot
plt.figure(figsize=(12, 8))  # Adjust figure size for better spacing
sns.boxplot(data=filtered_df, x=' Label', y=' Total Fwd Packets', palette="viridis")
plt.title('Total Forward Packets by Label', fontsize=16)
plt.xlabel('Label', fontsize=12)
plt.ylabel('Total Forward Packets', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

# PIE CHARTS

```python
import seaborn as sns
import matplotlib.pyplot as plt

# PIE CHART
label_counts = df['Label'].value_counts()

# Calculate percentages
percentages = label_counts / label_counts.sum() * 100

# Filter out labels with less than 5% representation
filtered_labels = percentages[percentages >= 5].index
filtered_counts = label_counts[filtered_labels]

# Create the pie chart
plt.figure(figsize=(8, 8))  # Adjust figure size for better visualization
patches, texts, autotexts = plt.pie(
    filtered_counts,
    labels=filtered_labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("viridis", len(filtered_counts)),
    wedgeprops={'edgecolor': 'black', 'linewidth': 0.5}  # Added edge color and line width
)

# Improve aesthetics of the text
for text in texts:
    text.set_fontsize(10)
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color('white')

# Add a legend outside the pie chart
plt.legend(patches, filtered_labels, title="Labels", loc="center left", bbox_to_anchor=(1, 0.5))

# Add title and adjust layout
plt.title('Distribution of Labels', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# PIE CHART for Protocols
protocol_counts = df['Protocol'].value_counts()

# Calculate percentages for Protocols
protocol_percentages = protocol_counts / protocol_counts.sum() * 100

# Filter out protocols with less than 5% representation
filtered_protocols = protocol_percentages[protocol_percentages >= 5].index
filtered_protocol_counts = protocol_counts[filtered_protocols]

# Create the pie chart for Protocols
plt.figure(figsize=(8, 8))  # Adjust figure size for better visualization
patches, texts, autotexts = plt.pie(
    filtered_protocol_counts,
    labels=filtered_protocols,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("magma", len(filtered_protocol_counts)),
    wedgeprops={'edgecolor': 'black', 'linewidth': 0.5}  # Added edge color and line width
)

# Improve aesthetics of the text
for text in texts:
    text.set_fontsize(10)
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color('white')

# Add a legend outside the pie chart for Protocols
plt.legend(patches, filtered_protocols, title="Protocols", loc="center left", bbox_to_anchor=(1, 0.5))

# Add title and adjust layout
plt.title('Distribution of Protocols', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
