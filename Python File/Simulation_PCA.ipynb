```python
from xgboost import XGBClassifier
```

```python
lb=joblib.load('/Users/mayankkapadia/Desktop/SJSUMATERIAL/SEM1/DV/Project/Packages and Models/labelencoded.pkl')
predict=lb.inverse_transform(predict_encoded)
```

```python
import requests
import json

# Power BI API URL and headers
powerbi_url = "https://api.powerbi.com/beta/a8eec281-aaa3-4dae-ac9b-9a398b9215e7/datasets/0cdd00dd-b8bd-4a2d-ba42-9ecec642e215/rows?experience=power-bi&key=aUgX%2FuvkMXQWakdX0eSsUQp5dqsbC7Bmo7yu6gu8HUiT0ui0Ltv8a1QAFdqnt%2FJlzPbhJt8mEjhs4yeHLKZLzw%3D%3D"
headers = {"Content-Type": "application/json"}

# Loop through each row and send as a single data object
for _, data in df_scaled.iterrows():
    # Prepare payload for a single row
    payload = {
        column: data[column] for column in df_scaled.columns
    }
    # Add predicted label
    payload["Predict_Label"] = predict[_]

    # Convert to JSON format as a list of one dictionary (required by Power BI)
    response = requests.post(powerbi_url, headers=headers, data=json.dumps([payload]))

    # Check the response
    if response.status_code in [200, 202]:
        print(f"Row {_} sent successfully to Power BI.")
    else:
        print(f"Failed to send row {_}: {response.status_code}, {response.text}")
```
